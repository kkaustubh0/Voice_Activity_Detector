```python
import soundfile as sf
import librosa
import numpy as np
from vad import EnergyVAD
from IPython.display import Audio

# Load audio file
audio_file = "/OSR_us_000_0010_8k.wav"
waveform, sample_rate = librosa.load(audio_file, sr=16000, mono=True)

# Reshape waveform to have 1 channel
waveform = waveform.reshape(1, -1)

# Create VAD instance
vad = EnergyVAD(
    sample_rate=sample_rate,
    frame_length=25,  # in milliseconds
    frame_shift=20,   # in milliseconds
    energy_threshold=0.05,  # you may need to adjust this value
    pre_emphasis=0.95,
)

# Get voice activity
voice_activity = vad(waveform)

# Apply VAD to get audio file with only speech segments
speech_signal = vad.apply_vad(waveform)

# Save audio file with only speech segments
speech_file = "/speech_segments.wav"
sf.write(speech_file, speech_signal.flatten(), sample_rate)
print("Playing audio with only speech segments...")
Audio(data=speech_signal.flatten(), rate=sample_rate)
```

```python
# Play both the original audio and the audio with only speech segments
print("Playing original audio...")
Audio(data=waveform.flatten(), rate=sample_rate)
```
